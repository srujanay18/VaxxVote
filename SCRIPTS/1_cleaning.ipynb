# Read in the vaccination datasets for Rubella, DTP, and Influenza

```python
# read all the vaccination datasets and display their first 4 rows
rubella_df = pd.read_excel("/content/Rubella vaccination coverage 2025-20-10 09-42 UTC.xlsx")
dtp_df = pd.read_excel("/content/Diphtheria tetanus toxoid and pertussis (DTP) vaccination coverage 2025-15-10 15-11 UTC (2).xlsx")
influenza_df = pd.read_csv("/content/Influenza Vaccine Doses Distributed in the United States, By Season (1).csv")

print("Rubella Head:")
display(rubella_df.head())

print("DTP Head:")
display(dtp_df.head())

print("Influenza Head:")
display(influenza_df.head())
```

# Cleaning Process Part 1 - drop the columns unneeded for analysis/model: TARGET_NUMBER and DOSES. Do this for the Rubella and DTP datasets.

```python
# drop the unneccesary columns for analysis: TARGET_NUMBER and DOSES
rubella_df = rubella_df.drop(columns=['TARGET_NUMBER', 'DOSES'])
dtp_df = dtp_df.drop(columns=['TARGET_NUMBER', 'DOSES'])

print("Rubella Head after dropping columns:")
display(rubella_df.head())

print("DTP Head after dropping columns:")
display(dtp_df.head())
```

# Cleaning Process Part 2 - drop the rows/observations where COVERAGE_CATEGORY == 'WUENIC' and COVERAGE_CATEGORY_DESCRIPTION == 'WHO/UNICEF Estimates of National Immunization Coverage'. Do this for the Rubella and DTP.

```python
# filter to only show observations where COVERAGE_CATEGORY == 'WUENIC' and COVERAGE_CATEGORY_DESCRIPTION == 'WHO/UNICEF Estimates of National Immunization Coverage'
rubella_df_filtered = rubella_df[(rubella_df['COVERAGE_CATEGORY'] == 'WUENIC') & (rubella_df['COVERAGE_CATEGORY_DESCRIPTION'] == 'WHO/UNICEF Estimates of National Immunization Coverage')]
dtp_df_filtered = dtp_df[(dtp_df['COVERAGE_CATEGORY'] == 'WUENIC') & (dtp_df['COVERAGE_CATEGORY_DESCRIPTION'] == 'WHO/UNICEF Estimates of National Immunization Coverage')]

print("Filtered Rubella Head:")
display(rubella_df_filtered.head())

print("Filtered DTP Head:")
display(dtp_df_filtered.head())
```

# Cleaning Process Part 3 - combine the cleaned Rubella, DTP, and Measles datasets into one dataframe

```python
# combine the cleaned Rubella, DTP, and Measles datasets into one dataframe
merged_vaccines_df = rubella_df_filtered.merge(dtp_df_filtered, on='YEAR', how='outer', suffixes=('_Rubella', '_DTP'))

display(merged_vaccines_df.head())
```

```python
vaccination_dfs = [rubella_df, dtp_df]

combined_vaccination_df = pd.concat(vaccination_dfs, ignore_index=True)

# Filter to keep only rows where COVERAGE_CATEGORY is 'WUENIC'
combined_vaccination_df = combined_vaccination_df[combined_vaccination_df['COVERAGE_CATEGORY'] == 'WUENIC'].copy()

# Filter to keep only rows with '1st dose' in ANTIGEN_DESCRIPTION
combined_vaccination_df = combined_vaccination_df[combined_vaccination_df['ANTIGEN_DESCRIPTION'].str.contains('1st dose')].copy()


combined_vaccination_df.dropna(inplace=True)

combined_vaccination_df = combined_vaccination_df.rename(columns={'YEAR': 'year'})

combined_vaccination_df['year'] = combined_vaccination_df['year'].astype(int)

display(combined_vaccination_df.head())
display(combined_vaccination_df.info())
```

# Cleaning Process Part 4 - due to influenza having different columns and looking at dosage only, combine the Influenza dataset with the presidents dataframe.

```python
# combine the Influenza dataset with the presidents dataframe
influenza_df['year'] = influenza_df['Date'].apply(lambda x: int(x.split('-')[0]))
merged_influenza_df = pd.merge(influenza_df, presidents_df, on='year', how='inner')

print("Merged Influenza Data Head:")
display(merged_influenza_df.head())
```

# Cleaning Process Part 5 - due to influenza having different columns and looking at dosage only, combine the Rubella, DTP, and Measles merged dataframe with the presidents dataframe.

```python
# combine the Rubella, DTP, and Measles merged dataframe with the presidents dataframe
merged_df = pd.merge(combined_vaccination_df, presidents_df, on='year', how='inner')
display(merged_df.head())
```

# Cleaning Process Part 6 - download the final dataframes as csv files


```python
# download the 3 vaccinations and presidents datasets as a csv file
merged_df.to_csv('merged_vaccination_presidents_data.csv', index=False)
```

```python
# download the influenza and presidents datasets as a csv file
merged_influenza_df.to_csv('merged_influenza_presidents_data.csv', index=False)
```
